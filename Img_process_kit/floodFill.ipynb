```python
import cv2
import matplotlib.pyplot as plt

def plt_show(img_ndarray, figsize=(8,6), gray=True):
    plt.figure(figsize=figsize)
    plt.grid(False)
    if gray:
        plt.imshow(img_ndarray, 'gray')
    else:
        plt.imshow(img_ndarray)
    plt.show()
```

```python
img = cv2.imread("./notedata/filled.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, th1 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt_show(th1)
```

```python
# 反转th1, 让前景为白色，背景为黑色
th2 = cv2.bitwise_not(th1)
plt_show(th2)
```

```python
# 填充函数
def fillHole(srcBw):
    temp = srcBw.copy()
    cv2.floodFill(temp, None, (0,0), 255)
    #temp = cv2.bitwise_not(temp)
    return temp

out = fillHole(th2)
plt_show(out)
```
